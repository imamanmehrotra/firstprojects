```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
y = data['price']
X = data.iloc[:,:-1]

```

```python
num_pipeline = Pipeline(
    steps = [
        ("imputer",SimpleImputer()),
        ("scalar", StandardScaler())
    ]

)

```

```python

cat_pipeline = Pipeline(
steps = 
[
    ("imputer", SimpleImputer(strategy='most_frequent')),
    ("ordinalencoder",OrdinalEncoder(categories=[cut_categories, color_categories, clarity_categories]))
]

)
```

```python
cut_categories = ['Fair', 'Good', 'Very Good','Premium','Ideal']
color_categories = ['D', 'E', 'F', 'G', 'H', 'I', 'J']
clarity_categories = ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
Preprocessor = ColumnTransformer(
[
 ("Numerical Pipeline", num_pipeline,num_cols),
 ("Cat Pipeline", cat_pipeline,cat_cols)
]

)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train_scaled = Preprocessor.fit_transform(X_train)
```

```python
X_test_scaled = Preprocessor.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train_scaled, columns = Preprocessor.get_feature_names_out())
X_test = pd.DataFrame(X_test_scaled, columns=Preprocessor.get_feature_names_out())
```

```python
X_test
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, ElasticNet, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
models = {

    'Linear Regression':LinearRegression(),
    "Lasso Regression": Lasso(),
    "Ridge Regression" : Ridge(),
    "Elastic Net": ElasticNet(),
    "XG Boost": XGBRegressor()
}
```

```python
nr = X_train.shape[0]
nf = X_train.shape[1]
nr, nf
```

```python
def evaluate_model(true, pred, no_of_features, no_of_rows):
    r2 = r2_score(true, pred)
    adj_r2 = 1 - ((1 - r2)*(no_of_rows - 1))/(no_of_rows-no_of_features-1)
    mse = mean_squared_error(true, pred)
    rmse = mse**0.5
    mae = mean_absolute_error(true, pred)

    return mae, mse,r2,rmse, adj_r2
```

```python
train_model_list = []
model_list = []
r2_list = []
rmse_list = []
adj_r2_list=[]


for selected_model in models.values():
    # print(selected_model)
    selected_model.fit(X_train, y_train)
    y_pred = selected_model.predict(X_test)

    MAE, MSE, R2 ,RMSE, adj_r2 = evaluate_model(y_test, y_pred, nf, nr)

    print(f"Model testing performance for {selected_model}")
    print("MAE:",MAE)
    print("MSE:",MSE)
    print("R2 Score:",R2)
    print("Adj R2 Score:",adj_r2)
    print("RMSE:",RMSE)

    model_list.append(selected_model)
    r2_list.append(R2)
    adj_r2_list.append(adj_r2)
    rmse_list.append(RMSE)


    print("*"*80)
    print("\n")
```

```python
df = pd.DataFrame([model_list, r2_list, adj_r2_list,rmse_list]).T
df.columns = [['Model',"R2", "Adj_R2","RMSE"]]
```
